```python
import pandas as pd

df = pd.read_csv("./training_data.csv", encoding='cp1252')

df.head()
```

```python
X = df[["MENGE_FREI_SHOWBEGINN", "ONAIR_SEKUNDEN"]].values
Y = df[["SOLD_OUT_FLG"]].values
```

```python
# KFold Cross Validation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

import numpy as np


kf = KFold(n_splits = 10, shuffle = True)
scores = cross_val_score(LinearRegression(), X, Y, cv = kf)

print(scores)
print(np.mean(scores))
```

```python
# Repeated KFold Cross Validation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

import numpy as np


kf = KFold(n_splits = 10, shuffle = True)
scores = cross_val_score(LinearRegression(), X, Y, cv = RepeatedKFold(n_splits = 10, n_repeats = 10))

print(scores)
print(np.mean(scores))
```
