```python
import pandas as pd

df = pd.read_csv(r"D:\google下载\1_training_data.csv", encoding='cp1252',engine = "python")


df.drop(["AIRING_23_FLG", "AIRING_456_FLG",'BESTELL_ID','WGH1_ID','WGH3_ID','WGH4_ID','DIVISION_DESC_SORT','WGH4_DESC','BEWERTUNG','FARBE'],axis=1,inplace=True)

df.head()
```

```python
df['SHOW_DATUM']=pd.to_datetime(df['SHOW_DATUM'])
df['DAY_OF_WEEK'] = df['SHOW_DATUM'].dt.day_name()
df["DAY_OF_WEEK"] = df["DAY_OF_WEEK"].astype('category')
df["DAY_OF_WEEK"] = df["DAY_OF_WEEK"].cat.codes



df['SHOW_DATUM']=pd.to_datetime(df['SHOW_DATUM'])
df['DAY_OF_WEEK'] = df['SHOW_DATUM'].dt.day_name()
df['Hour_of_the_day'] = df['SHOW_DATUM'].dt.hour
df.Hour_of_the_day
df.drop('SHOW_DATUM', axis = 1, inplace = True)
df.DAY_OF_WEEK.replace("Sunday", 0, inplace = True)
df.DAY_OF_WEEK.replace("Monday", 1, inplace = True)
df.DAY_OF_WEEK.replace("Tuesday", 2, inplace = True)
df.DAY_OF_WEEK.replace("Wednesday", 3, inplace = True)
df.DAY_OF_WEEK.replace("Thursday", 4, inplace = True)
df.DAY_OF_WEEK.replace("Friday", 5, inplace = True)
df.DAY_OF_WEEK.replace("Saturday", 6, inplace = True)
df.DAY_OF_WEEK.unique()
df.DAY_OF_WEEK.astype('category', inplace = True)
def from_hour_to_time_slot(x):
    if 7 <= x < 9:
        return "MORNING TIME"
    elif 9 <= x <17:
        return "DAYTIME"
    elif 17 <= x < 20:
        return "EARLY PRIME TIME"
    elif 20 <= x < 23:
        return "PRIME TIME"
    elif 1 <= x < 7:
        return "LATE NIGHT"
    else:
        return "LATE PRIME TIME"

df.Hour_of_the_day = df.Hour_of_the_day.map(from_hour_to_time_slot)
df.Hour_of_the_day.unique()
df.Hour_of_the_day.replace("MORNING TIME", 0, inplace = True)
df.Hour_of_the_day.replace("DAYTIME", 1, inplace = True)
df.Hour_of_the_day.replace("EARLY PRIME TIME", 2, inplace = True)
df.Hour_of_the_day.replace("PRIME TIME", 3, inplace = True)
df.Hour_of_the_day.replace("LATE PRIME TIME", 4, inplace = True)
df.Hour_of_the_day.replace("LATE NIGHT", 5, inplace = True)
df.Hour_of_the_day.astype('category', inplace = True)
```

```python

df['PREISKLASSE_DESC']=df['PREISKLASSE_DESC'].str[:2]

df['PREISKLASSE_DESC']=df['PREISKLASSE_DESC'].convert_objects(convert_numeric=True)
```

```python

df["PREIS_LABEL_DESC"] = df["PREIS_LABEL_DESC"].astype('category')
df["PREIS_LABEL_DESC"] = df["PREIS_LABEL_DESC"].cat.code
df["WG_DESC"] = df["WG_DESC"].astype('category')
df["WG_DESC"] = df["WG_DESC"].cat.codes
df["WGH1_DESC"] = df["WGH1_DESC"].astype('category')
df["WGH1_DESC"] = df["WGH1_DESC"].cat.codes
df["GROESSE"] = df["GROESSE"].astype('category')
df["GROESSE"] = df["GROESSE"].cat.codes
```

```python

def trans_wgh(x):
    if x =="Accessoire":
        return 0
    elif x == "Shirts":
        return 1
    elif x == "Jacken":
        return 2
    elif x == "Outdoor Jacken":
        return 3
    elif x == "Hosen":
        return 4
    elif x == "Strickware":
        return 5
    elif x == "Blusen":
        return 6
    elif x == "Westen":
        return 7
    elif x == "Kleider":
        return 8
    elif x == "Shirtware":
        return 9
    elif x == "Blazer":
        return 10
    elif x == "Röcke":
        return 11
    elif x == "Sweatware":
        return 12
    elif x == "Mäntel":
        return 13
    elif x == "Fleece-Ware":
        return 14
    elif x == "Homewear":
        return 15

df.WGH1_DESC = df.WGH1_DESC.map(trans_wgh)
df.WGH1_DESC = df.WGH1_DESC.astype("category")
```
