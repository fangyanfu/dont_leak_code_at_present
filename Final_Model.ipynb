```python
# classifies the different observations for the specific Hour_of_the_day depending 
# on the time into six different clusters
def from_hour_to_time_slot(x):
    if 7 <= x < 9:
        return "MORNING TIME"
    elif 9 <= x <17:
        return "DAYTIME"
    elif 17 <= x < 20:
        return "EARLY PRIME TIME"
    elif 20 <= x < 23:
        return "PRIME TIME"
    elif 0 <= x < 7:
        return "LATE NIGHT"
    else:
        return "LATE PRIME TIME"
# maps the data in the merged dataset to the clusters created in the function
result.Hour_of_the_day = result.Hour_of_the_day.map(from_hour_to_time_slot)

# converts the feature Hour_of_the_day into a category variable
result["Hour_of_the_day"] = result["Hour_of_the_day"].astype('category')
result["Hour_of_the_day"] = result["Hour_of_the_day"].cat.codes
result["Hour_of_the_day"] = result["Hour_of_the_day"].astype('category')

# drops the featurw SHOW_DATUM as all valuable information has to be extracted into new features
result.drop(['SHOW_DATUM'],axis=1,inplace=True)
```

```python
# fills the missing values of MENGE_ECOM with a value of zero
result['MENGE_CALL']=result['MENGE_CALL'].fillna(0)

# defines the features that should be used in order to predict the feature MENGE_CALL used in the training data
features = result[["MENGE_ECOM", "MENGE_FREI_SHOWBEGINN", "ONAIR_SEKUNDEN", "ADT_FLG", "BRAND_FLG", "PREIS_DISCOUNT",
                  'FAKTOR', 'ADT_FLG']]

# defines the feature that should be predicted
target= result[["MENGE_CALL"]]

# defines the features that should be used in order to predict the feature MENGE_CALL used in the test data
predict= result[["MENGE_ECOM", "MENGE_FREI_SHOWBEGINN", "ONAIR_SEKUNDEN", "ADT_FLG", "BRAND_FLG", "PREIS_DISCOUNT",
                  'FAKTOR', 'ADT_FLG']]

# shows the shape of the target and the "input" parameters for the Regression Model
print(features.shape)
print(target.shape)
```

```python
# imports necessary librariers / models from scikitlearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# sets up the Standard Scaler
scaler = StandardScaler()

# splits the data into training and test data
X, X_test, y, y_test = train_test_split(features, target, test_size = 0.2, random_state = 42)
# fits the training data
scaler.fit(X)

# transforms the training and test data
X = scaler.transform(X)
X_test = scaler.transform(X_test)
predict = scaler.transform(predict)

# shows the shape of the training and test data of the Regression Model
print(X.shape)
print(X_test.shape)
print(y.shape)
print(y_test.shape)
```

```python
# defines the Linear Regression Model
LR = LinearRegression()
# fits the training data to the model
LR.fit(X, y)
# prints the score of the test and the training data (shows how many percent of the dataset was predicted correctly)
print(LR.score(X_test, y_test))
print(LR.score(X, y)) 

# predicts the target values based on the test data
y_pred = LR.predict(X_test)
```

```python
# specifying the dataset to be the x in the function corr_df
x = results

# creates the function corr_df
def corr_df(x, corr_val):
    '''
    Obj: Drops features that are strongly correlated to other features.
          This lowers model complexity, and aids in generalizing the model.
    Inputs:
          df: features df (x)
          corr_val: Columns are dropped relative to the corr_val input (e.g. 0.6)
    Output: df that only includes uncorrelated features
    '''

    # Creates Correlation Matrix and Instantiates
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterates through Correlation Matrix Table to find correlated columns
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = item.values
            if val >= corr_val:
                # Prints the correlated feature set and the corr val
                print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(i)

    drops = sorted(set(drop_cols))[::-1]

    # Drops the correlated columns
    for i in drops:
        col = x.iloc[:, (i+1):(i+2)].columns.values
        df = x.drop(col, axis=1)

    return df
# definition of the correlation boarder, meaning that features that have a correlation of 0.60 or higher
# will be dropped
results = corr_df(results, 0.60)
```

```python
# imports the library to plot a confusion matrix
from sklearn.metrics import confusion_matrix
import itertools

# formula to calculate the confusin matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        #print("Normalized confusion matrix")
    else:
        1#print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
